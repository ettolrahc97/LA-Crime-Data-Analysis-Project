```python
# CREATE NEW COLUMN TO STORE THE DIFFERENCE BETWEEN date_reported and date_occurred_no_time
df['Time To Report'] = pd.to_datetime(df['date_reported']) - pd.to_datetime(df['date_occurred_no_time'])
```

```python
# Count the time to report
df['Time To Report'].value_counts().head(10)
```

```python
# CONVERTING  THE TIME TO REPORT TO DAYS
df['Time To Report'] = (pd.to_datetime(df['date_reported']) - pd.to_datetime(df['date_occurred_no_time'])).dt.days
```

```python
# TAKE THE AVERAGE OF  TIME TO REPORT

df.groupby(['crime_description'])['Time To Report'].mean().sort_values().head(10)

#  This shows the crimes how quickly takes times to report
```
